```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca data dari CSV
data = pd.read_csv('prochoice_prolife.csv')

# Ubah kolom 'created_at' menjadi tipe datetime
data['created_at'] = pd.to_datetime(data['created_at'])

# Sort data berdasarkan 'created_at'
data = data.sort_values(by='created_at')
plt.figure(figsize=(12, 6))
plt.plot(data['created_at'], data.index, marker='o', linestyle='-')
plt.title('Grafik Line Plot berdasarkan created_at')
plt.xlabel('Tanggal')
plt.ylabel('Index Data')
plt.grid(True)
plt.show()
```

### Kesimpulan 
#### Pada data set prochoice_prolife.csv, grafik line plot menunjukkan peningkatan yang signifikan dalam aktivitas atau atribut yang diukur dari bulan 6 tanggal 23 hingga bulan 7 tanggal 03. Periode ini menunjukkan peningkatan data yang signifikan, yang ditunjukkan dengan kenaikan tajam pada plot. Ini menunjukkan bahwa selama periode waktu tersebut, terjadi perubahan atau peristiwa signifikan yang mempengaruhi karakteristik yang diamati.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca data dari CSV
data = pd.read_csv('prochoice_prolife.csv')

# Hitung jumlah retweet untuk setiap entri data
retweet_counts = data['retweet_count'].value_counts()

# Ambil 10 nilai teratas (atau sesuaikan sesuai kebutuhan)
top_retweets = retweet_counts.head(10)

# Plot grafik batang
plt.figure(figsize=(12, 6))
top_retweets.plot(kind='bar', color='salmon')
plt.title('Grafik Batang Retweet Count')
plt.xlabel('Jumlah Retweet')
plt.ylabel('Jumlah Data')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

## Kesimpulan 
### Dalam data set prochoice_prolife.csv, ada grafik batang yang menunjukkan jumlah Retweet Count yang diterima oleh entri-entri dalam data; sebagian besar entri memiliki jumlah Retweet Count yang relatif rendah, dan hanya sedikit entri yang memiliki jumlah Retweet Count yang signifikan. Ini menunjukkan bahwa beberapa posting mungkin sangat populer dan mendapatkan banyak retweet, tetapi mayoritas posting memiliki dampak yang lebih kecil. Analisis tambahan dapat membantu dalam pemahaman lebih lanjut tentang komponen yang memengaruhi tingkat interaksi dan retweet dalam data ini, serta menemukan posting-posting yang mungkin memiliki dampak yang lebih besar dalam kaitannya dengan topik yang dibahas.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca data dari CSV
data = pd.read_csv('prochoice_prolife.csv')

# Pastikan kolom 'kategori' ada dalam data Anda
if 'kategori' in data.columns:
    # Group data berdasarkan 'kategori' dan hitung jumlah like_count
    kategori_like = data.groupby('retweet_count')['like_count'].sum()
    
    # Urutkan kategori berdasarkan jumlah like_count secara menurun
    kategori_like = kategori_like.sort_values(ascending=False)
    
    # Ambil 5 kategori teratas (atau sesuaikan sesuai kebutuhan)
    top_kategori = kategori_like.head(5)
    
    # Buat pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(top_kategori, labels=top_kategori.index, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'lightpink'])
    plt.title('Pie Chart Kategori Posting berdasarkan Like Count')
    plt.axis('equal')  # Agar pie chart menjadi lingkaran
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca data dari CSV
data = pd.read_csv('prochoice_prolife.csv')

# Hitung jumlah entri untuk setiap kategori atau tujuan posting
category_counts = data['public_metrics'].value_counts()
# Ambil 5 kategori teratas (atau sesuaikan sesuai kebutuhan)
top_categories = category_counts.head(5)

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_categories, labels=top_categories.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Kategori Posting berdasarkan Public Metrics')
plt.axis('equal')  # Agar pie chart terlihat seperti lingkaran
plt.show()
```

## Kesimpulan
### Dengan menggunakan pie chart pada data set prochoice_prolife.csv yang menggambarkan persentase public_metrics, kita dapat melihat sebaran kategori atau tujuan posting dalam dataset tersebut. Proporsi masing-masing kategori atau tujuan terlihat jelas pada grafik ini, yang memungkinkan kita untuk dengan cepat menemukan kategori yang paling dominan dalam dataset tersebut. Analisis lebih lanjut dapat membantu memahami fokus atau perbandingan antara berbagai jenis posting yang berkaitan dengan metrik publik. Hasil analisis juga dapat membantu dalam membuat keputusan dan strategi.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca data dari file CSV
data = pd.read_csv('prochoice_prolife.csv')

# Pilih atribut Retweet Count dan Like Count
atribut = data[['retweet_count', 'like_count']]

# Menghitung korelasi antar-atribut
korelasi = atribut.corr()

# Membuat heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(korelasi, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi antara Retweet Count dan Like Count')
plt.show()
```

## Kesimpulan 
### Heatmap di atas menunjukkan korelasi antara jumlah like dan retweet dalam dataset prochoice_prolife.csv. Korelasi adalah ukuran seberapa erat dua sifat berhubungan satu sama lain. Dalam hal ini, heatmap berfungsi sebagai alat untuk menunjukkan korelasi antara seberapa sering postingan di-retweet dan seberapa sering mereka mendapatkan like. Jika titik data pada heatmap berwarna merah atau kuning, itu menunjukkan korelasi positif, yang berarti postingan yang mendapatkan lebih banyak retweet juga cenderung mendapatkan lebih banyak like, dan sebaliknya, jika titik data berwarna biru atau hijau, itu menunjukkan korelasi negatif. Oleh karena itu, heatmap ini membantu memahami pola interaksi antara like dan retweet dalam dataset tersebut.
